```python
import pandas as pd
import random
```

```python
l = []
for i in range(10000):
    # l.append(random.randint(1,6))
    a = random.randint(1,6)
    b = random.randint(1,6)
    l.append(a+b)
    
             
```

```python
len(l)      
```

```python
l[:5]
```

```python
s = (pd.Series(l).value_counts()/pd.Series(l).value_counts().sum()).sort_index()
s
```

```python
s.plot(kind = 'bar')
```
