```python
class Count:
  def __init__(self, max):
    self.max = max
    self.current = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self.current<self.max:
      self.current+=1
      return self.current
    else:
      raise StopIteration

```

```python
class squares:
  def __init__(self, n):
    self.n= n
    self.current = 1

  def __iter__(self):
    return self

  def __next__(self):
    if self.current<= self.n:
      value = self.current**2
      self.current+=1
      return value
    else:
      raise StopIteration


s = squares(5)
print(next(s))
print(next(s))
print(next(s))
```

```python
# You are given a list of integers.
# Return:

# The first non-repeating number

# If none exists, return None

def non_repeat(number):

  count = {}

  for num in number:
    if num in count:
      count[num]+=1
    else:
      count[num]=1

  for num in number:
    if count[num]==1:
      return num

  return None

number = [1,2,3,2,1,4,3]
print(non_repeat(number))
```

```python
# Given a list of integers, return all elements that appear exactly once, in original order.
def non_repeating(number):
  count = {}
  non_re_num = []
  for num in number:
    if num in count:
      count[num]+=1
    else:
      count[num]=1

  for num in number:
    if count[num]==1:
      non_re_num.append(num)

  return non_re_num

number = [1,2,3,1,4,5,2,3]
print(non_repeating(number))


```

```python
# You receive mixed prediction outputs from a faulty pipeline.

# Input:
# preds = [1, 0, "a", None, 1, 1, "x", 0, 1]

# Task:

# Ignore invalid values (non-int)

# Compute accuracy %

# Count invalid entries

def mixed_predictions(pred):
  valid = 0
  invalid = 0
  correct = 0

  for num in pred:

    if isinstance(num, int):
      valid+=1
      if num==1:
        correct+=1
      else:
        invalid+=1

  accuracy = round((correct/valid)*100,2)
  return accuracy, invalid

pred = [1, 0, "a", None, 1, 1, "x", 0, 1]
print(mixed_predictions(pred))

```

```python
# You receive prediction probabilities.

# Input:
# probs = [0.9, 0.2, "x", 0.8, None, 0.4, 0.95]

# Rules:

# Ignore invalid values

# Convert to prediction:

# >= 0.5  1

# < 0.5  0

# Return:

# (correct_predictions, total_valid)

def pred_prob(pred):
  correct = 0
  total_valid = 0

  for p in pred:
    if isinstance(p, float):
      total_valid+=1
      if p>=0.5:
        correct+=1

  return correct, total_valid

pred = [0.9, 0.2, "x", 0.8, None, 0.4, 0.95]
print(pred_prob(pred))
```
