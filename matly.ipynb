```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa' : 0, 'Iris-virginica': 1, 'Iris-versicolor': 2}) 
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c= iris['Species'],cmap = 'summer', alpha = 0.7)
plt.xlabel('SepalLength')
plt.ylabel('PetaLength')
plt.colorbar()
```

```python
#PLOT SIZE

plt.figure(figsize = (15,7))
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c= iris['Species'],cmap = 'summer', alpha = 0.7)
plt.xlabel('SepalLength')
plt.ylabel('PetaLength')
plt.colorbar()
```

```python
#Annotations
batters = pd.read_csv('batter.csv')
batters.head(5)
```

```python
df_sample = batters.head(100).sample(25, random_state = 5)
df_sample
```

```python
plt.figure(figsize= (18,10))
plt.scatter(df_sample['avg'], df_sample['strike_rate'], df_sample['runs'])

#horizontal and vertical lines on graph
plt.axhline(130, color='red')
plt.axvline(30, color='red')
for i in range(df_sample.shape[0]):
    plt.text(df_sample['avg'].values[i], df_sample['strike_rate'].values[i], df_sample['batter'].values[i])
```
