```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv("Placed_Students.csv")
dataset.head(4)
```

```python
plt.figure(figsize = (4,3))
sns.scatterplot(x ='cgpa', y= 'resume_score', data = dataset, hue = 'placed')
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y= dataset['placed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import Perceptron
```

```python
p = Perceptron(alpha=0.1)
p.fit(x_train, y_train)
```

```python
p.score(x_train, y_train)*100,p.score(x_test, y_test)*100
```

```python
plt.figure(figsize= (4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(),clf = p)
plt.show()
```
