```python
import numpy as np
import matplotlib.pyplot as plt

# Values of X (number of heads)
x = np.array([0, 1, 2])

# PMF values
pmf = np.array([0.25, 0.5, 0.25])

# CDF: cumulative sum of PMF
cdf = np.cumsum(pmf)

# Plotting
plt.figure(figsize=(12, 5))

# PMF plot
plt.subplot(1, 2, 1)
plt.stem(x, pmf, basefmt=" ")
plt.title("PMF - Coin Toss (2 times)")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")

# CDF plot
plt.subplot(1, 2, 2)
plt.step(x, cdf, where='mid', marker='o', color='green')
plt.title("CDF - Coin Toss (2 times)")
plt.xlabel("Number of Heads")
plt.ylabel("Cumulative Probability")

plt.tight_layout()
plt.show()
```

```python
#pmf cdf

x = ([1,2,3,4,5,6])
pmf = np.full_like(x, 1/6, dtype= float)

cdf = np.cumsum(pmf)
plt.figure(figsize = (12,5))

plt.subplot(1,2,1)
plt.stem(x, pmf,basefmt = " ")
plt.title("PMF of die roll")
plt.xlabel("outcome")
plt.ylabel("Prob")

plt.subplot(1,2,2)
plt.step(x, cdf, where = 'mid', marker = 'o', color = 'green')
plt.title("CDF of x")

plt.tight_layout()
plt.show()
```

```python
sample_mean = 98
population_mean = 100
population_std = 5
n = 30
alpha = 0.05

# Step 2: Calculate z-score
z_cal = (sample_mean - population_mean) / (population_std / np.sqrt(n))
print(f"Calculated Z = {z_cal:.2f}")
```

```python
# Step 3: Find z-critical (left-tailed)
z_table = st.norm.ppf(alpha)  # notice: for left-tailed, no 1-alpha
print(f"Z-table value (critical) = {z_table:.2f}")
```

```python
# Step 4: Decision
if z_cal < z_table:
    print("Reject Null Hypothesis (Battery lasts less!)")
else:
    print("Fail to Reject Null Hypothesis (Claim is correct)")
```
