```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#regularization
ds = pd.read_csv(r"Housing.csv")
ds.head(5)
```

```python
numeric_ds = ds.select_dtypes(include=[np.number])
```

```python
plt.figure(figsize= (10,10)
sns.heatmap(data=numeric_ds.corr(), annot = True)
plt.show()
```

```python
ds.drop(['id','date'], axis = 1, inplace = True)
```

```python
x = ds.drop('price', axis = 1)
y = ds['price']
```

```python
sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x), columns = x.columns)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

LinearRegression

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)*100
```

```python
print(mean_squared_error(y_test, lr.predict(x_test)))
print(mean_absolute_error(y_test, lr.predict(x_test)))
print(np.sqrt(mean_squared_error(y_test, lr.predict(x_test))))
```

```python
lr.coef_
```

```python
plt.figure(figsize = (15,5))
plt.bar(x.columns, lr.coef_)
plt.title("Linear Regression")
plt.xlabel("columns")
plt.ylabel("coeff")
plt.show()
```

Lasso

```python
la = Lasso(alpha=0.5)
la.fit(x_train, y_train)
la.score(x_test, y_test)*100
```

```python
print(mean_squared_error(y_test, la.predict(x_test)))
print(mean_absolute_error(y_test, la.predict(x_test)))
print(np.sqrt(mean_squared_error(y_test, la.predict(x_test))))
```

```python
plt.figure(figsize = (15,5))
plt.bar(x.columns, la.coef_)
plt.title("Lasso")
plt.xlabel("columns")
plt.ylabel("coeff")
plt.show()
```

Ridge

```python
ri = Ridge(alpha = 10)
ri.fit(x_train, y_train)
ri.score(x_test, y_test)*100
```

```python
print(mean_squared_error(y_test, ri.predict(x_test)))
print(mean_absolute_error(y_test, ri.predict(x_test)))
print(np.sqrt(mean_squared_error(y_test, ri.predict(x_test))))
```

```python
plt.figure(figsize = (15,5))
plt.bar(x.columns, ri.coef_)
plt.title("Ridge")
plt.xlabel("columns")
plt.ylabel("coeff")
plt.show()
```

```python
compare_coef = pd.DataFrame({"columns": x.columns, "Linear_Regression": lr.coef_, "Lasso": la.coef_, "ridge": ri.coef_})
compare_coef
```
